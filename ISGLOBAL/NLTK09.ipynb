#### NUBE DE PALABRAS

```python
import pandas as pd
```

```python
df=pd.read_excel("CUALIT.xlsx")
```

```python
df.columns
```

```python
texto = " ".join(df["¿Si pudiera hacer algún cambio lo haría?"])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

nltk.download('stopwords')
nltk.download('punkt')

# Obtener las palabras de la cadena de texto
palabras = word_tokenize(texto.lower())

# Eliminar las palabras comunes
palabras_sin_stopwords = [palabra for palabra in palabras if palabra not in stopwords.words('spanish')]

# Obtener la frecuencia de cada palabra
frecuencia_palabras = FreqDist(palabras_sin_stopwords)

# Crear el gráfico de nube de palabras
plt.figure(figsize=(12,6))
frecuencia_palabras.plot(8, cumulative=False)
plt.show()
```
