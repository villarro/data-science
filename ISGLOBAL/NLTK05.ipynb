```python
import pandas as pd
from textblob import TextBlob
```

```python
# Leer el archivo de Excel
df = pd.read_excel('CUALIT.xlsx')
```

```python
df.head(2)
```

```python
# Seleccionar las tres columnas con texto a analizar
columnas_texto = ['¿Les fueron de utilidad los materiales de comunicación?',
                  '¿Considera qué hay algún impacto en los pacientes/comunidad? ¿Cuál? ',
                  '¿Si pudiera hacer algún cambio lo haría?']
```

```python
columnas_texto
```

```python
# Crear una nueva columna con el sentimiento asociado para cada columna de texto en español
for columna in columnas_texto:
    df[columna+'_sentimiento'] = df[columna].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Guardar los resultados en un nuevo archivo de Excel
df.to_excel('ruta_del_archivo_con_sentimiento.xlsx', index=False)
```
